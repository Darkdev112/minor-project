```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
```

```python
import pandas as pd
dataset1 = pd.read_csv('train_values.csv')
dataset2 = pd.read_csv('result_data_modified4.csv')
dataset2_multiplied = dataset2 * 2
merged_dataset = pd.concat([dataset1, dataset2_multiplied], axis=1)
merged_dataset.to_csv('merged_dataset.csv', index=False)
```

```python
merged_dataset
```

```python
X = merged_dataset[['aqi']]
y = merged_dataset['Value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=28, random_state=42)
```

```python
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)
```

```python
predictions = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", mse)
```

```python
def predict_crop_yield(row):
    input_data = row[['aqi']]
    input_data = input_data.values.reshape(1, -1).astype('float32')
    prediction = model.predict(input_data)
    return prediction[0][0]
```

```python
test_data['predicted_yield'] = test_data.apply(predict_crop_yield, axis=1)
```

```python
for index, row in test_data.iterrows():
    print("AQI:", row['aqi'], "- Predicted Crop Yield:", row['predicted_yield'])
```
