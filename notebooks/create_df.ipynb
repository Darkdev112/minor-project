```python
import os
import xarray as xr
import numpy as np
```

```python
given_lat= -90
given_lon= 179.9
```

```python
data = xr.open_dataset('../data/sample.nc4')
data
```

```python
conc_data = data["XCO2"].values
lat_data = data.coords["lat"].values
lon_data = data.coords["lon"].values
date_data = data.coords['time'].values
```

```python
conc = np.array(conc_data)[0]
lat = np.round(np.array(lat_data),1)
lon = np.round(np.array(lon_data),10)
conc
```

```python
date_val = date_data[0]
```

```python
date_column = np.empty((0,),dtype='datetime64')
date_column = np.append(date_column, date_val)
date_column
```

```python
lon_index = int(((np.floor(given_lon/0.625) * 0.625) - lon[0])/0.625)
lon_index = 0 if lon_index == 576 else lon_index
lat_index = int(((np.floor(given_lat/0.5) * 0.5) - lat[0])/0.5)
lon_index
```

```python
conc_val = conc[lat_index][lon_index]
```

```python
conc_column = np.empty((0,),dtype='float64')
conc_column = np.append(conc_column,conc_val)
conc_column
```
